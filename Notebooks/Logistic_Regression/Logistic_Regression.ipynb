```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
# Create dataset in a way that we dont have to do standardisation
```

```python
# Create datadset
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(x) # independent features
```

```python
pd.DataFrame(y) # dependent features
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred = logistic.predict(x_test)
print(y_pred)
```

```python
logistic.predict_log_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
score = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
p_score = precision_score(y_test,y_pred)
r_score = recall_score(y_pred, y_test)
c_report = classification_report(y_test, y_pred)
print(score)
print(cm)
print(p_score)
print(r_score)
print(c_report)
```

### Hyperparameter Tuning and Cross Validation

### GridSeachCV

```python
model = LogisticRegression(max_iter=5000)
# penalty = ['l1', 'l2', 'elasticnet'] 
penalty = ['l2']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

| Penalty     | Supported Solvers                        |
|-------------|-------------------------------------------|
| `l1`        | `liblinear`, `saga`                      |
| `l2`        | `newton-cg`, `lbfgs`, `liblinear`, `sag`, `saga` |
| `elasticnet`| `saga`                                   |
| `none`      | `newton-cg`, `lbfgs`, `sag`, `saga`      |

```python
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
# GridSearch CV
from sklearn.model_selection import GridSearchCV
Grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
Grid
```

```python
Grid.fit(x_train,y_train)
```

```python
Grid.best_params_
```

```python
y_pred_grid = Grid.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
score = accuracy_score(y_test,y_pred_grid)
cm = confusion_matrix(y_test,y_pred_grid)
p_score = precision_score(y_test,y_pred_grid)
r_score = recall_score(y_test,y_pred_grid)
c_report = classification_report(y_test, y_pred_grid)
print(score)
print(cm)
print(p_score)
print(r_score)
print(c_report)
```
